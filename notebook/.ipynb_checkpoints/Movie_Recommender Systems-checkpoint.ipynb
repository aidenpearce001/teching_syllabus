```python
from ast import literal_eval

df = pd.read_csv("../dataset/movies/movies_metadata.csv") 
features = ['genres']

for feature in features:
    df[feature] = df[feature].apply(literal_eval)
def get_list(x):
    if isinstance(x, list):
        str1 = ""
        names = [i['name'] for i in x]
        if len(names) > 3:
            names = names[:3]
        for ele in names: 
            str1 += ele +','
        return str1

    return []

from collections import Counter

data_vis = {}
data_vis = pd.DataFrame(data_vis)
count = []
genres = []
df['genres'] = df['genres'].apply(get_list)
dup = df['genres'].tolist()
# Counter(dup)
dct = dict(Counter(dup))
for k,v in dct.items():
    print(k,v)
    genres.append(k)
    count.append(v)
data_vis['genres'] = genres
data_vis['count'] = count
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer

tfidf = TfidfVectorizer(stop_words='english')

df['overview'] = df['overview'].fillna('')

tfidf_matrix = tfidf.fit_transform(df['overview'])

tfidf_matrix.shape
```

```python
from sklearn.metrics.pairwise import linear_kernel

cosine_sim = linear_kernel(tfidf_matrix, tfidf_matrix)
indices = pd.Series(df.index, index=df['title']).drop_duplicates()
```

```python
def get_recommendations(title, cosine_sim=cosine_sim):
    idx = indices[title]

    sim_scores = list(enumerate(cosine_sim[idx]))

    sim_scores = sorted(sim_scores, key=lambda x: x[1], reverse=True)

    sim_scores = sim_scores[1:11]

    movie_indices = [i[0] for i in sim_scores]

    return df['title'].iloc[movie_indices]
```

```python
get_recommendations('Avenger')
```
